### Junções de dados CETESB + SUS + Meteorologia

```python
df_polu = pd.read_csv('Dataset-SUS-OFICIAL.csv', delimiter=',')
df_polu = df_polu.drop(columns=['Unnamed: 0'])
df_met = pd.read_csv('Dataset-METAR-OFICIAL.csv', delimiter=',')
df_met = df_met.drop(columns=['Unnamed: 0'])
df_sus = pd.read_csv('Dataset-CETESB-OFICIAL.csv', delimiter=';')
df_sus = df_sus.drop(columns=['Unnamed: 0'])
```

##### CETESB + SUS

```python
df = pd.merge(df_polu, df_sus, on='Data', how='left')
# df
```

```python
df = df.drop_duplicates(subset=['Data'])
# df
```

```python
# em alguns dados de saúde acabaram ficando alguns NaNs. Esse código é só p/ add 0 nesses casos

df = df.fillna(value=0)
# df
```

##### (CETESB + SUS) + METAR

```python
df_total = pd.merge(df, df_met, on='Data', how='left')
df_total
```

#### Caso 1

Salvando o arquivo final usado no trabalho.

```python
df_total = df_total.to_csv('Dataset-REAL-OFICIAL-1.csv', sep=',')
```

#### Caso 3
ST longa (caso 1) + coluna como n° de casos de Covid 19.

```python
df_total = pd.read_csv('Dataset-REAL-OFICIAL-1.csv', delimiter=',')
df_total = df_total.drop(columns=['Unnamed: 0'])
df_total = df_total.drop(columns=['Unnamed: 0.1'])

df_covid = pd.read_csv('dados_covid_sp.csv', delimiter=';')
# df_covid = df_total.drop(columns=['Unnamed: 0'])
# df_covid
```

```python
df_covid_sp = df_covid.query('nome_munic == "São Paulo"')
df_covid_sp = df_covid_sp.reset_index(drop = True)
df_covid_sp["datahora"] = pd.to_datetime(df_covid_sp["datahora"])
# df_covid_sp
```

```python
atributos = list(df_covid_sp.columns)

atributos.remove("casos_novos")
atributos.remove("datahora")
```

```python
df_sp_filtrado = df_covid_sp.drop(columns = atributos )
df_sp_filtrado = df_sp_filtrado.rename(columns = {'datahora': 'Data'})
df_sp_filtrado
```

```python
df_tot_cov = pd.merge(df_total, df_sp_filtrado, on='Data', how='left')
```

```python
df_tot_cov = df_tot_cov.fillna(value = 0)
df_tot_cov
```

```python
df_tot_cov.to_csv("Dataset-REAL-OFICIAL-Covid-1.csv", sep = ',')
```

#### Caso 4
Casos diários acumulados

```python
df = pd.read_csv('Dataset-REAL-OFICIAL-1.csv', delimiter=',')
df = df.drop(columns=['Unnamed: 0'])
df = df.drop(columns=['Unnamed: 0.1'])
```

```python
casos = df['Casos Diários'].values
casos
```

```python
casos_cumulativos = np.cumsum(casos)
casos_cumulativos
```

```python
df = df.drop(columns=['Casos Diários'])
```

```python
df['Casos Diários Cumulativos'] = casos_cumulativos
df
```

```python
# salvando em um .csv

df.to_csv('Dataset-REAL-OFICIAL-Cumulativo.csv', sep=',')
```
