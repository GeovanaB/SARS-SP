```python
import pandas as pd
import os

# Colunas que você realmente precisa
colunas_desejadas = [
    "DT_NOTIFIC",   # Data da notificação
    "CS_SEXO",      # Sexo
    "DT_NASC",      # Data de nascimento
    "CLASSI_FIN"    # Classificação final
]

anos = range(2008, 2025)  # de 2008 até 2024
output_file = "INFLUD_2008_2024.csv"

# Remove o arquivo de saída se já existir
if os.path.exists(output_file):
    os.remove(output_file)

primeiro = True  # controla cabeçalho no CSV

# Dicionário para acumular frequência de notificações por dia
freq_total = {}

for ano in anos:
    try:
        sufixo = str(ano)[-2:].zfill(2)
        nome_arquivo = f"INFLUD{sufixo}.csv"

        print(f"Processando {nome_arquivo}...")

        # Lê em pedaços
        chunks = pd.read_csv(
            nome_arquivo,
            sep=";",
            on_bad_lines="skip",
            encoding="latin1",
            usecols=lambda c: c in colunas_desejadas,
            chunksize=100000
        )

        for chunk in chunks:
            chunk["ANO"] = ano

            # Converte datas
            chunk["DT_NOTIFIC"] = pd.to_datetime(chunk["DT_NOTIFIC"], errors="coerce")

            # Conta notificações por dia nesse pedaço
            freq = chunk["DT_NOTIFIC"].value_counts().to_dict()

            # Atualiza o dicionário global
            for data, count in freq.items():
                if pd.isna(data):
                    continue
                freq_total[data] = freq_total.get(data, 0) + count

            # (Ainda não adiciona a coluna "NOTIF_DIA", isso será feito depois)
            chunk.to_csv(
                output_file,
                mode="a",
                sep=";",
                index=False,
                encoding="utf-8-sig",
                header=primeiro
            )
            primeiro = False

    except FileNotFoundError:
        print(f"Arquivo {nome_arquivo} não encontrado. Pulando...")
    except ValueError:
        print(f"Algumas colunas de {nome_arquivo} não existem. Pulando esse ano...")

print("✅ Dataset bruto criado em:", output_file)

# Agora recarregamos o CSV final e adicionamos a coluna de notificações por dia
print("Adicionando coluna de frequência diária...")
df_final = pd.read_csv(output_file, sep=";", encoding="utf-8-sig")

df_final["DT_NOTIFIC"] = pd.to_datetime(df_final["DT_NOTIFIC"], errors="coerce")
df_final["NOTIF_DIA"] = df_final["DT_NOTIFIC"].map(freq_total)

# Salva de novo com a coluna extra
df_final.to_csv("INFLUD_2008_2024.csv", index=False, sep=";", encoding="utf-8-sig")

print("✅ Coluna NOTIF_DIA adicionada com sucesso no CSV final!")
```

```python
import pandas as pd
import os

colunas_desejadas = [
    "DT_NOTIFIC",   # Data da notificação
    "CS_SEXO",      # Sexo
    "DT_NASC",      # Data de nascimento
    "SG_UF_NOT",    # Estado de notificação
    "ID_MUNICIP",   # Município
    "EVOLUCAO",     # Evolução do caso
    "CLASSI_FIN"    # Classificação final
]

anos = range(2008, 2025)
output_file = "INFLUD_2008_2024.csv"

if os.path.exists(output_file):
    os.remove(output_file)

primeiro = True
freq_total = {}

for ano in anos:
    try:
        sufixo = str(ano)[-2:].zfill(2)
        nome_arquivo = f"INFLUD{sufixo}.csv"

        print(f"Processando {nome_arquivo}...")

        # Verifica quais colunas existem no arquivo
        colunas_arquivo = pd.read_csv(
            nome_arquivo,
            sep=";",
            nrows=0,
            encoding="latin1"
        ).columns.tolist()

        # Interseção
        colunas_para_ler = [c for c in colunas_desejadas if c in colunas_arquivo]
        if not colunas_para_ler:
            print(f"Nenhuma coluna de interesse encontrada em {nome_arquivo}. Pulando...")
            continue

        chunks = pd.read_csv(
            nome_arquivo,
            sep=";",
            on_bad_lines="skip",
            encoding="latin1",
            usecols=colunas_para_ler,
            chunksize=100000
        )

        for chunk in chunks:
            # Adiciona colunas faltantes como NaN
            for col in colunas_desejadas:
                if col not in chunk.columns:
                    chunk[col] = pd.NA

            chunk["ANO"] = ano

            # Converte datas
            chunk["DT_NOTIFIC"] = pd.to_datetime(
                chunk["DT_NOTIFIC"], errors="coerce", dayfirst=True
            )

            # Conta notificações por dia
            freq = chunk["DT_NOTIFIC"].value_counts().to_dict()
            for data, count in freq.items():
                if pd.isna(data):
                    continue
                freq_total[data] = freq_total.get(data, 0) + count

            # Garante ordem das colunas antes de salvar
            chunk = chunk[colunas_desejadas + ["ANO"]]

            chunk.to_csv(
                output_file,
                mode="a",
                sep=";",
                index=False,
                encoding="utf-8-sig",
                header=primeiro
            )
            primeiro = False

    except FileNotFoundError:
        print(f"Arquivo {nome_arquivo} não encontrado. Pulando...")

print("✅ Dataset bruto criado em:", output_file)

# Agora adiciona coluna NOTIF_DIA
print("Adicionando coluna de frequência diária...")
df_final = pd.read_csv(output_file, sep=";", encoding="utf-8-sig")
df_final["DT_NOTIFIC"] = pd.to_datetime(df_final["DT_NOTIFIC"], errors="coerce", dayfirst=True)
df_final["NOTIF_DIA"] = df_final["DT_NOTIFIC"].map(freq_total)

df_final.to_csv(output_file, index=False, sep=";", encoding="utf-8-sig")
print("✅ Coluna NOTIF_DIA adicionada com sucesso no CSV final!")
```

```python
import pandas as pd

# Ler o dataset (exemplo para 2009)
df = pd.read_csv("INFLUD09.csv", sep=";", encoding="latin1", on_bad_lines="skip")

# Converter coluna de data
df["DT_NOTIFIC"] = pd.to_datetime(df["DT_NOTIFIC"], errors="coerce", dayfirst=True)

# Tratar idade
# Em alguns datasets pode ser 'NU_IDADE_N' ou similar
if "NU_IDADE_N" in df.columns:
    df["IDADE"] = pd.to_numeric(df["NU_IDADE_N"], errors="coerce")
else:
    df["IDADE"] = None  # caso não tenha a coluna

# Classificação por faixa etária
def faixa_etaria(idade):
    if pd.isna(idade):
        return None
    elif idade <= 11:
        return "Criança"
    elif idade <= 17:
        return "Adolescente"
    elif idade <= 59:
        return "Adulto"
    else:
        return "Idoso"

df["FaixaEtaria"] = df["IDADE"].apply(faixa_etaria)

# Contagem de notificações por dia
freq_total = df.groupby("DT_NOTIFIC").size().reset_index(name="Freq_total")

# Contagem por faixa etária por dia (pivot)
freq_faixa = (
    df.groupby(["DT_NOTIFIC", "FaixaEtaria"]).size().unstack(fill_value=0).reset_index()
)

# Juntar tudo
result = pd.merge(freq_total, freq_faixa, on="DT_NOTIFIC", how="left")

# Renomear colunas para padronizar
result = result.rename(columns={
    "Criança": "Freq_crianca",
    "Adolescente": "Freq_adolescente",
    "Adulto": "Freq_adulto",
    "Idoso": "Freq_idoso"
})

print(result.head(20))
```
